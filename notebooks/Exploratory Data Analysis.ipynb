# Distribución LGD

```python
# Plot de la distribución de x14 (LGD)
plt.figure(figsize=(8, 6))
sns.histplot(df['x14'], kde=True, bins=100, color='blue')
plt.title('Distribución de LGD (x14)')
plt.xlabel('LGD (x14)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
descripcion_x14 = df['x14'].describe()

# Mostrar las estadísticas descriptivas
print("Estadísticas descriptivas para x14 (LGD):\n", descripcion_x14)
```

# Correlación

```python
dummies= pd.get_dummies(df[['x5', 'x10', 'x1', 'x2', 'x9', 'x3', 'x11', 'x4']])
X_2 = df.drop(['x5', 'x10', 'x1', 'x2', 'x9', 'x3', 'x11', 'x4', 'POPULATION'], axis=1).astype('float64')
```

```python
X_3 = pd.concat([X_2, dummies], axis=1)
```

```python
plt.figure(figsize=(22, 22))
sns.heatmap(X_3.corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Mapa de calor de correlaciones')
plt.show()
```

```python
correlaciones = X_3.corr()
print("Correlaciones entre las columnas del DataFrame:")
print(correlaciones)
```

# Distribución de variables categóricas 

```python
categorical_columns = df.select_dtypes(include=['object']).columns

plt.figure(figsize=(20, 20))

for i, col in enumerate(categorical_columns, 1):
    plt.subplot(3, 3, i)  
    sns.countplot(data=df, x=col, palette=sns.color_palette("flare"))  
    plt.title(f'Distribución de la variable categórica: {col}')  
    plt.xlabel(col) 
    plt.ylabel('Frecuencia')  

plt.tight_layout()  
plt.show()  
```

# Distribución variables binarias

```python
# Identificar variables binarias
binary_columns = [col for col in df.columns if df[col].nunique() == 2 and df[col].dtype in ['int64', 'float64']]
print("Variables binarias:")
print(binary_columns)
```

```python
plt.figure(figsize=(15, 15))

for i, col in enumerate(binary_columns, 1):
    plt.subplot(2, 2, i)  
    sns.countplot(data=df, x=col, palette=sns.color_palette("flare"))  
    plt.title(f'Distribución de la variable binaria: {col}')  
    plt.xlabel(col) 
    plt.ylabel('Frecuencia')  

plt.tight_layout()  
plt.show()  
```
